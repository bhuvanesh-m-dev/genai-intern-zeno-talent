```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-5, -3, -1, 0, 1, 3, 5])
```

```python
# Activation Function: Sigmoid
# Mathematical Formula:
# f(x) = 1 / (1 + e^(-x))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_sigmoid = sigmoid(x)
print("Sigmoid Output:", y_sigmoid)

plt.plot(x, y_sigmoid)
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
# Activation Function: Tanh
# Mathematical Formula:
# f(x) = (e^x - e^(-x)) / (e^x + e^(-x))

def tanh(x):
    return np.tanh(x)

y_tanh = tanh(x)
print("Tanh Output:", y_tanh)

plt.plot(x, y_tanh)
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
# Activation Function: ReLU
# Mathematical Formula:
# f(x) = max(0, x)

def relu(x):
    return np.maximum(0, x)

y_relu = relu(x)
print("ReLU Output:", y_relu)

plt.plot(x, y_relu)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
# Activation Function: Leaky ReLU
# Mathematical Formula:
# f(x) = max(0.01x, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

y_leaky_relu = leaky_relu(x)
print("Leaky ReLU Output:", y_leaky_relu)

plt.plot(x, y_leaky_relu)
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```
