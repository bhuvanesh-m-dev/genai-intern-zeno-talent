```python
import numpy as np
import matplotlib.pyplot as plt

# ----- Sample Data -----
y_true = np.array([2.5, 3.5, 4.5, 5.5])
y_pred = np.array([2.8, 3.0, 4.7, 5.0])

errors = y_true - y_pred
mse_points = (errors ** 2)
mae_points = np.abs(errors)
rmse_points = np.sqrt(mse_points)

x_index = np.arange(len(y_true))

# ----- MSE Plot -----
plt.figure()
plt.plot(x_index, mse_points)
plt.xlabel("Index")
plt.ylabel("MSE Contribution")
plt.title("Mean Squared Error (Per Sample)")
plt.grid(True)
plt.show()

# ----- MAE Plot -----
plt.figure()
plt.plot(x_index, mae_points)
plt.xlabel("Index")
plt.ylabel("MAE Contribution")
plt.title("Mean Absolute Error (Per Sample)")
plt.grid(True)
plt.show()

# ----- RMSE Plot -----
plt.figure()
plt.plot(x_index, rmse_points)
plt.xlabel("Index")
plt.ylabel("RMSE Contribution")
plt.title("Root Mean Squared Error (Per Sample)")
plt.grid(True)
plt.show()
```
