```python
student = {
    "student1": {"name": "Aarav",    "course": "Computer Science", "marks": 88},
    "student2": {"name": "Saanvi",   "course": "Mathematics",      "marks": 95},
    "student3": {"name": "Vihaan",   "course": "Physics",          "marks": 82},
    "student4": {"name": "Aaradhya", "course": "Biology",          "marks": 91},
    "student5": {"name": "Arjun",    "course": "Data Science",     "marks": 79},
    "student6": {"name": "Anaya",    "course": "English",          "marks": 85},
    "student7": {"name": "Reyansh",  "course": "Economics",        "marks": 90},
    "student8": {"name": "Diya",     "course": "Psychology",       "marks": 87},
    "student9": {"name": "Kabir",    "course": "Chemistry",        "marks": 93},
    "student10":{"name": "Zara",     "course": "History",          "marks": 89}
}

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
def assign_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "D"

for student_id, student_info in student.items(): # to traverse
    marks = student_info.get("marks")
    if marks is not None:
        student_info["grade"] = assign_grade(marks)

print("Updated student dictionary with grades:")
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
print("All keys and values in the student dictionary:")
for key, value in student.items():
    print(f"Key: {key}, Value: {value}")
```
