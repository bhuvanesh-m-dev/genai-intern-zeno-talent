```python
import datetime

def log_prompt():  # prompt from the user
    prompt_text = input("Enter your AI prompt: ")

    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")  # store time and date
    log_entry = f"[{timestamp}] {prompt_text}\n"

    try:
        with open("prompt_history.txt", "a") as f:
            f.write(log_entry)
        print("Prompt successfully saved to prompt_history.txt")
    except IOError as e:
        print(f"Error writing to file: {e}")

def read_history():
    try:
        with open("prompt_history.txt", "r") as f:
            content = f.read()
            if content.strip() == "":
                print("(File is empty)")
            else:
                print(content)
    except FileNotFoundError:
        print("prompt_history.txt does not exist yet.")

log_prompt()
read_history()
```
