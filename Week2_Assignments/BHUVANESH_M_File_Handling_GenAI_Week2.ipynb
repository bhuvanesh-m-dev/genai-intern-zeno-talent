```python
student_data = [
    "1    Aarav        Computer Science     88     B",
    "2    Saanvi       Mathematics          95     A",
    "3    Vihaan       Physics              82     B",
    "4    Aaradhya     Biology              91     A",
    "5    Arjun        Data Science         79     C"
]

with open("ai_students.txt", "a") as file:
    for student in student_data:
        file.write(student + "\n") # to write the student details

print("Student details have been written to ai_students.txt")
```

```python
print("Students who scored above 75 marks:")
with open("ai_students.txt", "r") as file:
    for line in file:
        parts = [part for part in line.strip().split(' ') if part]
        if len(parts) >= 4:
            try:
                # Marks at index 3
                marks = int(parts[3])
                if marks > 75:
                    # Print the full line
                    print(f"{parts[1]} scored {marks} marks in {parts[2]}")
            except (ValueError, IndexError):
                continue
```
