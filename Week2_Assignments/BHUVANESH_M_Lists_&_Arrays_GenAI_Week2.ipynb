```python
import random
random_integers_list = [] #to store the 10 numbers
for _ in range(10):
    random_int = random.randint(1, 100)#range of 1 to 100
    random_integers_list.append(random_int)
print(random_integers_list)
```

```python
print(f"Original list: {random_integers_list}")
random_integers_list.append(999)#append 999
print(f"List after adding 999: {random_integers_list}")

if 999 in random_integers_list:
    random_integers_list.remove(999)
    print(f"List after removing 999: {random_integers_list}")
else:
    print("999 was not found in the list to be removed.")

max_value = max(random_integers_list) #use max function
print(f"Maximum value in the list: {max_value}")


min_value = min(random_integers_list) #use min function
print(f"Minimum value in the list: {min_value}")

random_integers_list.sort()
print(f"List after sorting (ascending): {random_integers_list}") #use sort method
```

```python
import numpy as np
print(f"Original Python List: {random_integers_list}")

numpy_array = np.array(random_integers_list) #conert into from list to array
print(f"Converted NumPy Array: {numpy_array}")

mean_value = np.mean(numpy_array) #calculate mean
print(f"Mean of the array: {mean_value}")

median_value = np.median(numpy_array) #calculate median
print(f"Median of the array: {median_value}")

std_dev_value = np.std(numpy_array)
print(f"Standard Deviation of the array: {std_dev_value}")
```
