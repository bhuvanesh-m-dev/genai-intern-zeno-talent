```python
# @title Enter Numbers for Calculator
num1 = 0.0 # @param {type:"number", description:"Enter the first number"}
num2 = 0.0 # @param {type:"number", description:"Enter the second number"}

print(f"Numbers entered: {num1} and {num2}")
```

```python
num1,num2=34,21
print("Performing operations on ", num1, " and ", num2, ":")

print("Addition:", num1, "+", num2, "=", num1 + num2) #add
print("Subtraction:", num1, "-", num2, "=", num1 - num2)#sub
print("Multiplication:", num1, "*", num2, "=", num1 * num2)#mul

if num2 != 0:
    print("Division:", num1, "/", num2, "=", num1 / num2)#div
    print("Modulus:", num1, "%", num2, "=", num1 % num2)#mod
else:
    print("Division and Modulus by zero are not defined.")

print("Power:", num1, "**", num2, "=", num1 ** num2)#power
```

```python
print("Demonstrating Comparison and Logical Operators on ", num1, " and ", num2, ":")

# Comparison Operators
print("Comparison: ", num1, ">", num2, "is", num1 > num2)
print("Comparison: ", num1, "<", num2, "is", num1 < num2)
print("Comparison: ", num1, "==", num2, "is", num1 == num2)
print("Comparison: ", num1, "!=", num2, "is", num1 != num2)

# Logical Operators
print("Logical: (", num1, "> 20) and (", num2, "< 30) is", (num1 > 20) and (num2 < 30))
print("Logical: (", num1, "== 34) or (", num2, "== 10) is", (num1 == 34) or (num2 == 10))
print("Logical: not (", num1, "< 0) is", not (num1 < 0))
```

```python
print(f"Demonstrating formatted output with f-strings:")
print(f"The value of num1 is: {num1}")
print(f"The value of num2 is: {num2}")
print(f"A simple sum using f-string: {num1} + {num2} = {num1 + num2}")
```
